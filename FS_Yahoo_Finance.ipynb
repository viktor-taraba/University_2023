```python
import yfinance as yf
import pandas as pd
import lxml
import pandas as pd
from sqlalchemy import create_engine
import urllib
```

```python
quoted = urllib.parse.quote_plus("Driver={SQL Server Native Client 11.0};Server=VIKTOR-PC-2001;Database=DividendPolicyResearch;Trusted_Connection=yes;")
engine = create_engine('mssql+pyodbc:///?odbc_connect={}'.format(quoted))
result = engine.execute("SELECT [Ticker] FROM [dbo].[vw_tb_sp_tickers_with_CIK] where without_dividends = 0")
df_SP100_companies = pd.DataFrame(result.fetchall())
tickers_list = df_SP100_companies.iloc[:, 0].tolist()
tickers_list
```

```python
print(len(tickers_list))
```

```python
for ticker in tickers_list:
    ticker = ticker.replace("-",".")
    print("Тікер: " + ticker)
    
    request = yf.Ticker(ticker)
    df = request.financials
    df = df.transpose()
    df = df.filter(['Basic EPS', 'Diluted EPS'])

    df['ticker_name'] = ticker.replace(".","-")
    df['Date'] = df.index
    
    df.to_sql('tb_FS_YF', schema='dbo', con = engine, if_exists='append',index=False)
```
