```python
url = "https://in.investing.com/equities/apple-computer-inc-income-statement?period_type=annually"
#"https://in.investing.com/equities/dow-chemical-income-statement?periods=previous&period_type=quarterly&period_type=annually"
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Send a GET request to the specified URL
response = requests.get(url)
html = response.content

with open('output.html', 'w') as f:
    f.write(response.text)

#the HTML code you've written above
parsed_html = BeautifulSoup(html)

#with open("output.html", "w", encoding = 'utf-8') as file:
    # prettify the soup object and convert it into a string
  #  file.write(str(parsed_html.prettify()))
```

```python
for row in parsed_html.select('thead tr'):
    row_text = [x.text for x in row.find_all('th')]
    if row_text[0] == '\n\nName\n\n':
      if '\n' in row_text: row_text.remove('\n')
      if '\n\n\n ' in row_text: row_text.remove('\n\n\n ')
      if '' in row_text: row_text.remove('')
      row_text = list(map(lambda x: x.replace('\n', ''), row_text))
      #print(row_text)
      print(', '.join(row_text))
```

```python
fsli_list = [' Total Revenue',
' Revenue',
' Other Revenue, Total',
' Cost of Revenue, Total',
' Gross Profit',
' Total Operating Expenses',
' Selling/General/Admin. Expenses, Total',
' Research & Development',
' Depreciation / Amortization',
' Interest Expense (Income) - Net Operating',
' Unusual Expense (Income)',
' Other Operating Expenses, Total',
' Operating Income',
' Interest Income (Expense), Net Non-Operating',
' Gain (Loss) on Sale of Assets',
' Other, Net',
' Net Income Before Taxes',
' Provision for Income Taxes',
' Net Income After Taxes',
' Minority Interest',
' Equity In Affiliates',
' U.S GAAP Adjustment',
' Net Income Before Extraordinary Items',
' Total Extraordinary Items',
' Net Income',
' Total Adjustments to Net Income',
' Income Available to Common Excluding Extraordinary Items',
' Dilution Adjustment',
' Diluted Net Income',
' Diluted Weighted Average Shares',
' Diluted EPS Excluding Extraordinary Items',
' DPS - Common Stock Primary Issue',
' Diluted Normalized EPS]']

for row in parsed_html.select('tbody tr'):
    row_text = [x.text for x in row.find_all('td')]
    if row_text[0] in fsli_list:
      if '\n' in row_text: row_text.remove('\n')
      if '\n\n\n ' in row_text: row_text.remove('\n\n\n ')
      #print(row_text)
      print(', '.join(row_text))
```
